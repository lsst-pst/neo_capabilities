```python
%matplotlib inline
from __future__ import print_function, division
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FormatStrFormatter, FuncFormatter

plt.style.use(['ggplot', 'ggplot_mod.mplstyle'])
```

```python
tracklets_generated = [12890181, 25360438, 48483523, 125473320]
frac_nonasteroid_detections = [0.0, 49.7, 66.4, 79.8]
input_tracklets = [6312807, 13318186, 25824121, 66635397]
tracks_found = [285566, 306886, 343008, 461902]
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.plot(frac_nonasteroid_detections, np.array(tracklets_generated)/1e6, '-o', lw=2)
plt.ylabel("Number of Tracklets Generated")
plt.xlabel("Non-asteroid Sources in Detection Catalog")
plt.gca().xaxis.set_major_formatter(FormatStrFormatter("%.0f%%"))
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x,_: ("%.0f" % x) + "M"*int(x > 1)) )
plt.xlim(-5,85)

plt.subplot(1,2,2)
plt.plot(np.array(input_tracklets)/1e6, np.array(tracks_found)/1e3, '-o', lw=2)
#plt.gca().yaxis.set_major_formatter(FormatStrFormatter("%.0fk"))
plt.gca().xaxis.set_major_formatter(FormatStrFormatter("%.0fM"))
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x,_: ("%.0f" % x) + "k"*int(x > 1)) )

plt.xlim(xmin=1)
plt.ylim(ymin=0)
plt.ylabel("Number of Tracks Found")
plt.xlabel("Number of Input Tracklets")
plt.subplots_adjust(wspace=0.3)
plt.savefig("track_stats.pdf")
```
